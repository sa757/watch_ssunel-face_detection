```python
import cv2
import face_detection
import matplotlib.pyplot as plt
print(face_detection.available_detectors)
detector = face_detection.build_detector(
  "DSFDDetector", confidence_threshold=.5, nms_iou_threshold=.3, max_resolution=1080)
# BGR to RGB
im = cv2.imread("../baby.JPG")[:, :, ::-1].copy()
plt.imshow(im);plt.show()
detections = detector.detect(im)
#This will return a tensor with shape [N, 5], where N is number of faces and the five elements are [xmin, ymin, xmax, ymax, detection_confidence]
```

```python
for i in range(len(detections)):
    xmin, ymin, xmax, ymax = detections[i][:4].astype(int)
    cv2.rectangle(im, pt1=(xmin,ymin), pt2=(xmax,ymax), color=(0,0,255), thickness=2)
plt.imshow(im)
plt.show()
```
